```python
import pandas as pd
from bs4 import BeautifulSoup
import os
from IPython.display import clear_output
```

```python
# List all html files
files = []
for x in os.listdir():
    if x.endswith(".html"):
        # Prints only text file present in My Folder
        files.append(x)
```

```python
# Souping the files
def extract_data(soup):
    mydivs = soup.find_all("div", {"class": "rc-FormPartsQuestion"})
    #print("Number: ",len(soup.find_all('div',{"class":"rc-FormPartsQuestion__numberCell"}))//2)
    for i,div in enumerate(mydivs):
        question = div.find('div',{"data-test": 'legend'})
        answers = div.find_all('div',{"class":"rc-Option rc-Option--isReadOnly"})
        correct = []
        #print(f"Question {i+1}: ", end="")
        for q in question.select('p'):
            print(f"{q.text}")
        for j,answer in enumerate(answers):
            print(f"{chr(j+65)}: ",end="")
            for line in answer.select('p'):
                print(line.text)
            if answer.find("svg")["aria-labelledby"][:12] == "RadioChecked" or answer.find("svg")["aria-labelledby"][:8] == 'Checkbox':
                correct.append(answer.select('p'))
        print("---")
        for c in correct:
            for a in c:
                print(a.text)
        print("</>")
```

```python
# Printing the file
for file in files:
    with open(file,
              encoding='utf-8') as fp:
        soup = BeautifulSoup(fp, 'html.parser')
        extract_data(soup)
```
